```python
# 2. åŒ¯å…¥
from langchain_aws import BedrockLLM
from langchain.agents import initialize_agent, AgentType, Tool
from langchain.prompts import PromptTemplate

# 3. å»ºç«‹ Bedrock LLM
llm = BedrockLLM(
    model_id="anthropic.claude-3-sonnet-20240229-v1:0",  # çœ‹ä½ è¦ç”¨å“ªå€‹ï¼Œé€™æ˜¯ Claude 3 Sonnet
    model_kwargs={"temperature": 0.2},
)

# 4. è‡ªè¨‚ Promptï¼Œé€™æ˜¯é—œéµ
prompt = PromptTemplate.from_template(
    """ä½ æ˜¯ä¸€ä½å‹å–„ä¸”ç²¾æº–çš„åŠ©æ‰‹ï¼Œå°ˆé–€ä½¿ç”¨æä¾›çš„å·¥å…·ä¾†å›ç­”å•é¡Œã€‚
å¦‚æœä½ ä¸çŸ¥é“ç­”æ¡ˆï¼Œè«‹èª å¯¦å›ç­”ã€Œæˆ‘ä¸çŸ¥é“ã€ã€‚

å•é¡Œï¼š{input}"""
)

# 5. å¯ä»¥é¸æ“‡åŠ å·¥å…·ï¼ˆé€™è£¡å…ˆç©ºçš„ï¼‰
tools = []

# 6. åˆå§‹åŒ– agent
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    agent_kwargs={"input_variables": ["input"], "prompt": prompt},
    verbose=True,  # å¯ä»¥çœ‹åˆ°ä¸­é–“æ¨ç†éç¨‹
)

# 7. å‘¼å« agent
result = agent.run("å°ç£çš„é¦–éƒ½æ˜¯å“ªè£¡ï¼Ÿ")
print(result)
```
