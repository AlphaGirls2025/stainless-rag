```python
# 🔧 連接 AWS Bedrock
import boto3
from langchain_aws import BedrockLLM

# 建立 boto3 client
client = boto3.client(
    service_name="bedrock-runtime",
    region_name="us-east-1",  # ⚡ 換成你的 Bedrock 開通區域
)
```

```python
# 🚀 建立 LLM 對象
llm = BedrockLLM(
    model_id="anthropic.claude-v2",  # 👈 可以換別的，例如 titan-text-lite, jurassic-2
    client=client,
)
```

```python
# 📝 測試發送 Prompt
prompt = "請用中文寫一首關於春天的詩。"

response = llm.invoke(prompt)

print("=== 回應內容 ===")
print(response)
```

```python
# 2. 匯入
from langchain_aws import BedrockLLM
from langchain.agents import initialize_agent, AgentType, Tool
from langchain.prompts import PromptTemplate

# 3. 建立 Bedrock LLM
llm = BedrockLLM(
    model_id="anthropic.claude-3-sonnet-20240229-v1:0",  # 看你要用哪個，這是 Claude 3 Sonnet
    model_kwargs={"temperature": 0.2},
)

# 4. 自訂 Prompt，這是關鍵
prompt = PromptTemplate.from_template(
    """你是一位友善且精準的助手，專門使用提供的工具來回答問題。
如果你不知道答案，請誠實回答「我不知道」。

問題：{input}"""
)

# 5. 可以選擇加工具（這裡先空的）
tools = []

# 6. 初始化 agent
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    agent_kwargs={"input_variables": ["input"], "prompt": prompt},
    verbose=True,  # 可以看到中間推理過程
)

# 7. 呼叫 agent
result = agent.run("台灣的首都是哪裡？")
print(result)
```
